```python
# Box and Whisker Plots
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 10,10  # control plot size
df.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```
